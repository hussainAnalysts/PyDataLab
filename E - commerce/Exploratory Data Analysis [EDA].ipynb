##  Outlier Detection

```python
# Calculating Q1 and Q3 for 'price'
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]

outliers.head()
```

```python
 # Calculate Z-Score
# Calculating z-score for 'payment_value'
df['z_score_payment'] = (df['payment_value'] - df['payment_value'].mean()) / df['payment_value'].std()

# Identify extreme values (e.g., |z| > 3)
extreme_payments = df[df['z_score_payment'].abs() > 3]
extreme_payments.head()
```

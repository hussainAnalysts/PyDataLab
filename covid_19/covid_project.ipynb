# Covid 19 Data Analysis 

## 1.Data Cleaning 

```python
df.drop_duplicates()
```

```python
# Checking for missing values in each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# continent column
df['Continent'].fillna(df['Continent'].mode()[0], inplace=True)
df
```

```python
# population column 
df.dropna(subset=['Population'], inplace=True)
```

```python
# Cases columns
df['Cases.New'] = df['Cases.New'].fillna(0)
df['Cases.Active'] = df['Cases.Active'].fillna(0)
df['Cases.Critical'] = df['Cases.Critical'].fillna(0)
df['Cases.Recovered'] = df['Cases.Recovered'].fillna(0)
```

```python
df['Cases.1M_pop'].fillna(df['Cases.Total'] / df['Population'] * 1e6)
```

```python
# death per one million people
df['Cases.1M_pop'] = df['Cases.1M_pop'].fillna((df['Cases.Total'] / df['Population']) * 1e6)
```

```python
df['Deaths.New'] = df['Deaths.New'].fillna(0)
```

```python
# Deaths columns
df['Deaths.New'] = df['Deaths.New'].fillna(0)
df['Deaths.1M_pop'] = df['Deaths.1M_pop'].fillna((df['Deaths.Total'] / df['Population']) * 1e6)
df['Deaths.Total'] = df['Deaths.Total'].fillna(0)
```

```python
# Tests columns
df['Tests.1M_pop'] = df['Tests.1M_pop'].fillna(df['Tests.Total'] / df['Population'] * 1e6)
df['Tests.Total'] = df['Tests.Total'].fillna(0)
```

```python
# Verifing for any missing values,  if any missing values are left
print(df.isnull().sum())
```

```python
# Save the cleaned DataFrame to an Excel file
df.to_excel(r"C:\Users\user\Desktop\#\DA\practice\cleaned_covid19_data.xlsx", index=False)
```

```python
# Droping Unwanted columns
df.drop(['Day', 'Time'], axis=1, inplace=True)
```

## Visualizations

```python
import seaborn as sns
```

```python
# Bar chart for the top 10 countries by total cases
top_10_cases = mn.groupby('Country')['Cases.Total'].sum().nlargest(10)

# Adjusting figure size for smaller width and height
plt.figure(figsize=(10, 4))

# Reducing the width of the bars using the 'width' parameter
sns.barplot(x=top_10_cases.index, y=top_10_cases.values, palette="Blues_r", width=0.5)

# Adding labels and title
plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)

# Adding values on bars with smaller text size
for i, value in enumerate(top_10_cases.values):
    plt.text(i, value + 0.05 * value, f'{value:,.0f}', ha='center', va='bottom', fontsize=8)

# Adjusting the font size of the x-axis labels
plt.xticks(rotation=45, fontsize=10)

# Remove y-axis lines and ticks
plt.gca().yaxis.set_visible(False)  # Hides the y-axis
plt.tick_params(left=False)  # Removes y-axis ticks

# Remove the spines (lines surrounding the chart)
sns.despine()

plt.show()
```
